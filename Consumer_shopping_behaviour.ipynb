```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/shopping_behavior_updated.csv')
# pd.set_option('display.max_rows',20)
df
```

```python
df.describe()
```

```python
count1=df.groupby(['Category', 'Gender']).size().unstack(fill_value=0)
count1.plot.bar()
```

```python
filtered_df = df[df['Gender'] == 'Male']
filtered_df.groupby(['Item Purchased','Season']).size().unstack(fill_value=0).plot.bar(figsize=(15,10))
plt.xlabel('Category')
plt.ylabel('City')
plt.title('Categories in demand accross different city')
plt.legend()
plt.show()
```

```python
filtered_df = df[df['Gender'] == 'Female']
filtered_df.groupby(['Item Purchased','Season']).size().unstack(fill_value=0).plot.bar(figsize=(15,10))
plt.xlabel('Category')
plt.ylabel('City')
plt.title('Categories in demand accross different city')
plt.legend()
plt.show()
```

```python
max_goods=df.groupby(['Location', 'Category']).size().unstack(fill_value=0)
max_goods['sum']=max_goods.sum(axis=1)
max_goods.sort_values('sum', ascending=False, inplace=True)
```

```python
graph=max_goods.nlargest(10, 'sum')
```

```python
graph['sum'].plot.bar()
plt.xlabel('Top 10 cities')
plt.ylabel('Consumption')
plt.title('Goods consumption across top max consuming cities')
plt.legend()
plt.show()
```

```python
df1=df.groupby(['Purchase Amount (USD)', 'Category']).size().unstack(fill_value=0)
df1.reset_index(inplace=True)
df1
```

```python
df1['Total revenue']=df1['Purchase Amount (USD)']*(df1['Accessories']+df1['Clothing']+df1['Footwear']+df1['Outerwear'])
df1
```

```python
str(df1['Total revenue'].sum(axis=0)) + " USD"
```

```python
df1['Accessories']=df1['Purchase Amount (USD)']*df1['Accessories']
df1['Clothing']=df1['Purchase Amount (USD)']*df1['Clothing']
df1['Footwear']=df1['Purchase Amount (USD)']*df1['Footwear']
df1['Outerwear']=df1['Purchase Amount (USD)']*df1['Outerwear']
df2=df1.sum(axis=0)
```

```python
df2
```

```python
sum_table = pd.DataFrame(df2).transpose()
sum_table
```

```python
sum_table.columns
```

```python
sum_table[['Accessories', 'Clothing', 'Footwear','Outerwear']].plot.bar()
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.title('Revenue generated from each category')
plt.legend()
plt.show()
```
